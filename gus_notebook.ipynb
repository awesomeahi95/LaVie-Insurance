```python
import pandas as pd
```

```python
df = pd.read_csv('analytic_data2019.csv')
```

```python
df.head()
df.shape
```

```python
cicols = []
for column in df.columns:
    if ("CI" in column) or ("(Black)" in column) or ("(White)" in column) or ("(Hispanic)" in column) or ("numerator" in column) or ("denominator" in column):
        print(column)
        cicols.append(column)

df_clean = df.drop(cicols, axis=1)
```

```python
df_clean.isnull().sum()[df_clean.isnull().sum() > 100].sort_values(ascending = False)
```
